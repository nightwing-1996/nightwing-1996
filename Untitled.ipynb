```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('US police shootings in from 2015-22.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
camera=df.groupby('body_camera')['name'].count().sort_values().reset_index()
```

```python
camera.rename({'name':'count_of'},axis=1,inplace=True)
```

```python
camera
```

```python
plt.bar(camera['body_camera'],camera['count_of'])
plt.xticks([1,0])
plt.xlabel("0- no camera , 1-camera")
plt.ylabel("count of shoots")
plt.title("count of shoots with and without camera")
plt.show()
```

```python
df.head()
```

```python
shot=df.groupby('manner_of_death')['name'].count().sort_values(ascending=False).reset_index()
```

```python
shot.rename({'name':'count_of'},axis=1,inplace=True)
```

```python
plt.bar(shot['manner_of_death'],shot['count_of'])
plt.xlabel("manner of shot")
plt.ylabel("count of incidences")
plt.show()
```
