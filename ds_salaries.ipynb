```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('ds_salaries.csv',index_col='Unnamed: 0')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['salary','salary_currency'],inplace=True)
```

```python
df
```

```python
# salary by expiernce_level
```

```python
exp=df.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False).reset_index()
```

```python
exp
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(exp['experience_level'],exp['salary_in_usd'],color='red')
plt.ylabel("salary in k",fontsize=15)
plt.xlabel("exp level",fontsize=15)
plt.show()
```

```python
# salary by job title
```

```python
df.head()
```

```python
job=df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).reset_index().head(5)
```

```python
plt.figure(figsize=(13,15))
plt.bar(job['job_title'],job['salary_in_usd'])
```

```python
# count of company size
```

```python
size=df.groupby('company_size')['job_title'].count().sort_values(ascending=False).reset_index()
```

```python
plt.bar(size['company_size'],size['job_title'])
plt.xlabel("size of company")
plt.ylabel('count of companies')
plt.show()
```
